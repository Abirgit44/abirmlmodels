# Linear Regression

## importing necessary libraries

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Generating Data

```python
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + 2 * X**2 + np.random.randn(100, 1)
```

## Linear Regression class

```python
class LinearRegression:
    def __init__(self, learning_rate = 0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.theta = None
        
    def fit(self, X, y):
        m,n = X.shape
        self.theta = np.zeros((n+1,1))
        X_b = np.c_[np.ones((m,1)),X] #this is the bias item
        
        for _ in range(self.n_iterations):
            gradients = -2 / m * X_b.T.dot(X_b.dot(self.theta) - y)
            self.theta -= self.learning_rate * gradients
            
    def predict(self, X):
        m, n = X.shape
        X_b = np.c_[np.ones((m, 1)), X]
        return X_b.dot(self.theta)
```

## Fitting the model

```python
model = LinearRegression()
model.fit(X,y)
```

## Seeing predictions

```python
X_New = np.linspace(0, 2, 100).reshape(-1, 1)
y_pred = model.predict(X_New)
# Print the predicted values
print("Predictions for X_new:")
for i in range(len(X_New)):
    print(f"X_New: {X_New[i][0]}, Predicted y: {y_pred[i][0]}")
```

### Visualize the data and the linear regression line


```python
plt.scatter(X, y, alpha=0.6, label='Data')
plt.plot(X_New, y_pred, 'r-', label='Linear Regression', linewidth = 2)
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
